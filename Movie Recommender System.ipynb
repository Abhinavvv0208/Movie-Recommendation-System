```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# We have to bring the genres in this format-
# ['Action','Adventure','Fantasy']
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
       L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           L.append(i['name'])
           counter+=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
           L.append(i['name'])
           break 
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# Now we will convert the lines into list form

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# We will eliminate the spaces between the names
# like- Sam Wothington ---> SamWothington
# This process wil be done for genres,keywords,cast,crew
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
new_df= movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
# Here we will covert similar type of words into a single type
# ['loved','loving','love'] ---> ['love','love','love']
```

```python
def stem(text):
    y=[]

    for i in text.split():
       y.append( ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
cosine_similarity(vector).shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index= new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    return
    
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
